```python
import numpy as np
```

```python
X=np.linspace(0,100, num=2000)
X.shape=(2000,1)
```

```python
Y= 10+ 4.115*X+40*np.random.randn(2000,1)
```

```python
w0=(Y.mean()*(X**2).mean()-X.mean()*(X*Y).mean())/((X**2).mean()-X.mean()**2)
```

```python
print(w0)
```

```python
w1=((X*Y).mean()-X.mean()*Y.mean())/((X**2).mean()-X.mean()**2)
```

```python
print(w1)
```

```python
import matplotlib.pyplot as plt
```

```python
Ypred=w0+w1*X
```

```python
plt.scatter(X,Y)
plt.plot(X,Ypred, color="red")
```

```python
Rsquared=1-((Y-Ypred).T.dot(Y-Ypred))/((Y-Y.mean()).T.dot(Y-Y.mean()))
```

```python
print(Rsquared)
```
