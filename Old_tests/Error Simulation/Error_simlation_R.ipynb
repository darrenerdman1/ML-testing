```r
X=seq(from = 0, to = 100, by = 100/1999)
```

```r
length(X)
```

```r
runif(10,0,10)
```

```r
Y= 10+ 4.115*X+40*rnorm(2000,0,1)
```

```r
w0=(mean(Y)*mean(X**2)-mean(X)*mean(X*Y))/(mean(X**2)-mean(X)**2)
```

```r
w0
```

```r
w1=(mean(X*Y)-mean(X)*mean(Y))/(mean(X**2)-mean(X)**2)
```

```r
w1
```

```r
Ypred=w0+w1*X
```

```r
plot(X,Y,"p",col="blue",pch=19,cex=0.5)
lines(X,Ypred,"l", col="red", lwd=3)
```

```r
Rsquared=1-((Y-Ypred)%*%(Y-Ypred))/(Y-mean(Y))%*%(Y-mean(Y))
```

```r
Rsquared
```
