```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
N=1000
```

```python
X1=40*np.random.randn(N,2)
X2=47*np.random.randn(N,2)+150
```

```python
plt.scatter(X1[:,0],X1[:,1], color="blue", s=10)
plt.scatter(X2[:,0],X2[:,1], color="red",s=10)
```

```python
X1=np.hstack((X1,np.zeros((N,1))))
X2=np.hstack((X2,np.ones((N,1))))
```

```python
X=np.hstack((np.ones((2*N,1)),np.vstack((X1,X2))))
```

```python
X.shape
```

```python
np.random.shuffle(X)
```

```python
y=X[:,-1]
X=X[:,:3]
```

```python
y=y.reshape(2*N,1)
```

```python
def cost(y,p):
    for j in range(N):
        Err=[]
        if y[j]==1:
            Err.append(-np.log(p[j]))
        else:
            Err.append(-np.log(1-p[j]))
    return np.sum(Err)
```

```python
w=np.random.randn(3,1)
epochs=20000
eta=.000005
Error=[]
epoch=[]
```

```python
for i in range(epochs):
    p=1/(1+np.exp(-X@w))
    Error.append(cost(y,p))
    epoch.append(i)
    w=w-eta*X.T@(p-y)
    
```

```python
plt.scatter(np.array(epoch).reshape(epochs,1),Error[:])
```

```python
X2_pred=-(w[0]+w[1]*np.linspace(-120,250, num=2000))/w[2]
```

```python
plt.scatter(X1[:,0],X1[:,1], color="blue", s=6)
plt.scatter(X2[:,0],X2[:,1], color="red",s=6)
plt.plot(np.linspace(-120,250, num=2000),X2_pred, color="black", linewidth=5)
```
