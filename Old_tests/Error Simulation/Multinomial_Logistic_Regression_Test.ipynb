```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=1000
```

```python
X1=35*np.random.randn(N,1)
X2=45*np.random.randn(N,1)
```

```python
X=np.hstack((np.ones((N,1)),X1,X2))
```

```python
w_true=np.random.randn(3,4)
```

```python
def softmax(X,w):
    p=np.exp(X@w)
    return(p/(p.sum(1).reshape(N,1)))
```

```python
Y=np.rint(softmax(X,w))
```

```python
labels=np.argmax(Y,axis=1)
```

```python
colors=["red","blue","teal","black"]
```

```python
X2_true0=-(w_true[0,0]+w_true[1,0]*np.linspace(-125,125, num=2000))/w_true[2,0]
X2_true1=-(w_true[0,1]+w_true[1,1]*np.linspace(-125,125, num=2000))/w_true[2,1]
X2_true2=-(w_true[0,2]+w_true[1,2]*np.linspace(-125,125, num=2000))/w_true[2,2]
X2_true3=-(w_true[0,3]+w_true[1,3]*np.linspace(-125,125, num=2000))/w_true[2,3]
```

```python
plt.scatter(X[:,1],X[:,2], color=np.array(colors)[labels])
plt.plot(np.linspace(-125,125, num=2000),X2_true0, color="black", linewidth=5)
plt.plot(np.linspace(-125,125, num=2000),X2_true1, color="black", linewidth=5)
plt.plot(np.linspace(-125,125, num=2000),X2_true2, color="black", linewidth=5)
plt.plot(np.linspace(-125,125, num=2000),X2_true3, color="black", linewidth=5)
plt.axis([-125, 125, -150, 150])
```

```python
def cost(y,p):
    return -1*np.sum(y*np.log(p))
```

```python
w=np.random.randn(3,4)
epochs=20000
eta=.0001
Error=[]
epoch=[]
```

```python
for i in range(epochs):
    P=softmax(X,w)
    Error.append(cost(Y,P))
    epoch.append(i)
    w=w-eta*X.T@(P-Y)
    
```

```python
plt.scatter(epoch,Error)
```

```python
X2_pred0=-(w[0,0]+w[1,0]*np.linspace(-125,125, num=2000))/w[2,0]
X2_pred1=-(w[0,1]+w[1,1]*np.linspace(-125,125, num=2000))/w[2,1]
X2_pred2=-(w[0,2]+w[1,2]*np.linspace(-125,125, num=2000))/w[2,2]
X2_pred3=-(w[0,3]+w[1,3]*np.linspace(-125,125, num=2000))/w[2,3]
```

```python
plt.scatter(X[:,1],X[:,2], color=np.array(colors)[labels])
plt.plot(np.linspace(-125,125, num=2000),X2_true0, color="black", linewidth=5)
plt.plot(np.linspace(-125,125, num=2000),X2_true1, color="black", linewidth=5)
plt.plot(np.linspace(-125,125, num=2000),X2_true2, color="black", linewidth=5)
plt.plot(np.linspace(-125,125, num=2000),X2_true3, color="black", linewidth=5)
plt.plot(np.linspace(-125,125, num=2000),X2_pred0, color="red", linewidth=5)
plt.plot(np.linspace(-125,125, num=2000),X2_pred1, color="red", linewidth=5)
plt.plot(np.linspace(-125,125, num=2000),X2_pred2, color="red", linewidth=5)
plt.plot(np.linspace(-125,125, num=2000),X2_pred3, color="red", linewidth=5)
plt.axis([-125, 125, -150, 150])
```
