```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Z= np.linspace(-700,700,100)
```

```python
def tanh(X):
    return (np.exp(X)-np.exp(-X))/(np.exp(X)+np.exp(-X))
```

```python
plt.plot(Z,tanh(Z))
```

```python
X=np.random.randn(500,2)
```

```python
X=np.hstack((np.ones((500,1)),X))
```

```python
def sigmoid(X):
    return 1/(1+np.exp(-X))
```

```python
def layer_in(X,w):
    X=np.hstack((np.ones((X.shape[0],1)),X))
    return tanh(X@w)
```

```python
def layer_out(X,w):
    X=np.hstack((np.ones((X.shape[0],1)),X))
    return np.rint(sigmoid(X@w))
```

```python
w1=np.random.randn(3,6)
w2=np.random.randn(7,3)
w3=np.random.randn(4,1)
```

```python
out=layer_out(layer_in(layer_in(X,w1),w2),w3)
```

```python
colors=["blue","red"]
```

```python
plt.scatter(X[:,0],X[:,1], color=np.array(colors)[out.reshape(-1).astype(int)])
```

```python
w1=np.random.randn(2,6)
w2=np.random.randn(6,3)
w3=np.random.randn(3,1)
b1=np.random.randn(1,6)
b2=np.random.randn(1,3)
b3=np.random.randn(1,1)
```

```python
out=np.rint(sigmoid(tanh(tanh(X@w1+b1)@w2+b2)@w3+b3))
```

```python
plt.scatter(X[:,0],X[:,1], color=np.array(colors)[out.reshape(-1).astype(int)])
```

```python
boundary=xs[np.where(abs(tanh(tanh(xs@w1+b1)@w2+b2)@w3+b3)<.001)[0],:]
```

```python
plt.scatter(X[:,0],X[:,1], color=np.array(colors)[out.reshape(-1).astype(int)])
plt.scatter(boundary[:,0],boundary[:,1], color="black")
plt.ylim(-3, 3)
plt.xlim(-3,3)
```
