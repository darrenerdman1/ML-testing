```python
#Min max normalize
#X is data set to normalize, X_tr is data set to normalize bassed on
def norm(X,X_tr):
    X_tr_min=X_tr.min(0)
    X_tr_max=X_tr.max(0)
    X=(X-X_tr_min)/(X_tr_max-X_tr_min)
    return(X)
```
