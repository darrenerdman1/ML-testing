```python
import numpy as np
import matplotlib.pyplot as plt
import TSNN
```

```python
data=np.random.randn(5000,2)
data2=(25./100.)*np.random.randn(300,2)
data3=2.3*np.random.randn(10000,2)

circle=np.array([[x[0],x[1]] for x in data if (np.sqrt(x[0]**2+x[1]**2)>2.0)])
circle3=np.array([[x[0],x[1]] for x in data3 if (np.sqrt(x[0]**2+x[1]**2)>5.50)])

data2=np.hstack((data2,np.zeros((data2.shape[0],1))))
data=np.hstack((circle,np.ones((circle.shape[0],1))))
data3=np.hstack((circle3,2*np.ones((circle3.shape[0],1))))


data=data[:300,:]
data2=data2[:300,:]
data3=data3[:300,:]
X=np.vstack((data,data2,data3))
np.random.shuffle(X)
y=X[:,-1:]
X=X[:,:-1]
```

```python
colors=["red","blue","green"]
```

```python
X.shape
```

```python
plt.scatter([X[:,0]],[X[:,1]], color=np.array(colors)[y.reshape(-1).astype(int)], s=12)
```

```python
nodes=[20]*3+[3]
activs=['tanh']*3
```

```python
model=TSNN.NeuralNet(nodes,X.shape[1],activs,task="Classification",scaleweights=True)
```

```python
model.train(X,y,5000,1e-4,0)
```

```python
plt.plot(model.error)
```

```python
model.predict(X)
(model.predictions==y).mean()
```
